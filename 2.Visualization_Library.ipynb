```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 경고 무시
import warnings
warnings.filterwarnings(action='ignore')

# 그래프 그릴 때 이미지 품징을 더욱 선명하게 해주는 코드
%config InlineBackend.figure_format='retina'
```

```python
# 데이터 읽어오기
path = '/content/drive/MyDrive/DataForSudy/stock.csv'
stock = pd.read_csv(path)

# 전처리
stock = stock.loc[stock['Date'].between('2022-12-01', '2022-12-31')]
stock.reset_index(drop=True, inplace=True)
stock.insert(1, 'Day', stock['Date'].str.split('-').str[2])
stock.drop('Date', axis=1, inplace=True)

# 확인
stock.head()
```

```python
# 원하는 스타일로 그래프 그리기
plt.plot(stock['Day'], stock['Close'], linestyle='--', marker='o', color='r')
plt.title('Stock Closing Price', size=15, pad=10)
plt.xlabel('Day')
plt.ylabel('Close')
plt.show()
```

```python
# 여러 그래프 겹쳐 그리기
plt.plot(stock['Day'], stock['Open'], color='g', marker='o', linestyle='--')
# plt.plot(stock['Day'], stock['Open'], 'go--') -> 이와 같이 간단하게 수정 가능
plt.plot(stock['Day'], stock['Close'], color='r', marker='s', linestyle='--')
# plt.plot(stock['Day'], stock['Close'], 'rs--') -> 이와 같이 간단하게 수정 가능
plt.title('Stock Price', size=15, pad=10)
plt.xlabel('Day')
plt.ylabel('Price')
plt.show()
```

```python
# legend() 함수를 사용해 범례 표시
# label 속성을 사용해 각 그래프의 이름 지정
# grid 함수를 사용해 괘선 표시
plt.plot(stock['Day'], stock['Open'], 'o--', label='Open')
plt.plot(stock['Day'], stock['Close'], 's--', label='Close')
plt.title('Stock Price', size=15, pad=10)
plt.xlabel('Day')
plt.ylabel('Price')
plt.legend()
plt.grid() # axis 지정 가능 (both, x, y) default는 both
plt.show()
```

### matplotlib subplot 구조 예시

#### 1. 1행 2열 (1x2) subplot

|                   |                       |                       |
|:-----------------:|:---------------------:|:---------------------:|
|                   | **subplot(1,2,1)**    | **subplot(1,2,2)**    |

- `subplot(1, 2, index)` : 1행 2열, 오른쪽으로 나란히 두 개의 subplot

---

#### 2. 2행 1열 (2x1) subplot

|                   |                       |
|:-----------------:|:---------------------:|
|                   | **subplot(2,1,1)**    |
|                   | **subplot(2,1,2)**    |

- `subplot(2, 1, index)` : 2행 1열, 위아래로 두 개의 subplot

---

#### 3. 2행 2열 (2x2) subplot

|                |                 |                 |
|:--------------:|:---------------:|:---------------:|
|                | **subplot(2,2,1)** | **subplot(2,2,2)** |
|                | **subplot(2,2,3)** | **subplot(2,2,4)** |

- `subplot(2, 2, index)` : 2행 2열, 네 개의 subplot이 격자로 배치됨

---

> 💡 **index는 왼쪽 위(또는 맨 위)부터 1, 2, 3, ... 순서로 매겨집니다.**
>
> subplot 함수의 형태는 `subplot(nrows, ncols, index)`  
> (`nrows`=행의 개수, `ncols`=열의 개수, `index`=위치)

```python
# subplot을 통해 그래프 그리기

# 위 아래로 그래프 그리기
plt.subplot(2, 1, 1)
plt.plot(stock['Day'], stock['Open'], color='tab:blue', marker='o')
plt.title('Opening Price')

plt.subplot(2, 1, 2)
plt.plot(stock['Day'], stock['Close'], color='tab:orange', marker='s')
plt.title('Closing Price')

plt.tight_layout()
plt.show()
```

```python
# 옆으로 그리기
plt.subplot(1, 2, 1)
plt.plot(stock['Day'].astype(int), stock['Open'], color='tab:blue', marker='o')
plt.title('Opening Price')

plt.subplot(1, 2, 2)
plt.plot(stock['Day'].astype(int), stock['Close'], color='tab:orange', marker='s')
plt.title('Closing Price')

plt.tight_layout()
plt.show()
```

```python
# seaborn 라이브러리 불러오기
import seaborn as sns
```

```python
# 데이터 읽어오기
path = '/content/drive/MyDrive/DataForSudy/titanic_simple.csv'
titanic = pd.read_csv(path)
titanic.dropna(inplace=True)

#확인
titanic.head()
```

```python
# 데이터 읽어오기
path = '/content/drive/MyDrive/DataForSudy/iris.csv'
iris = pd.read_csv(path)

# 확인
iris.head()
```

```python
# histogram
sns.histplot(x='Age', data=titanic, bins=16, ec='k') # bins는 막대의 개수, ec는 edgecolor
plt.show()
```

```python
# hue를 사용래 특정 칼럼을 기준으로 구분
sns.histplot(x='Age', hue='Survived', data=titanic, bins=16, ec='k')
plt.show()
```

```python
# Density Plot
sns.kdeplot(x='Age', data=titanic)
plt.show()
```

```python
# common_norm=False를 지정하면 그래프 각각의 면적이 1
sns.kdeplot(x='Age', hue='Survived', data=titanic, common_norm=False)
plt.show()
```

```python
# common_norm=True를 지정하면 두 그래프 면적의 합이 1
sns.kdeplot(x='Age', hue='Survived', data=titanic, common_norm=True)
plt.show()
```

```python
# box plot
sns.boxplot(y='Age', data=titanic)
plt.show()
```

```python
# box plot
sns.boxplot(x='Age', data=titanic)
plt.show()
```

```python
# Survived 그룹 별로 Age의 분포를 가로 방향으로 나타냄
sns.boxplot(x='Age', y='Survived', data=titanic, orient='h')
plt.show()
```

```python
# DistributionPlot
sns.distplot(x=titanic['Age'], bins=16, hist_kws={'ec': 'k'})
plt.show()
```

```python
# JointPlot
sns.jointplot(x='Petal.Length', y='Petal.Width', data=iris)
plt.show()
```

```python
# PairPlot
sns.pairplot(iris, hue='Species')
plt.show()
```

```python
# CountPlot
sns.countplot(x='Embarked', data=titanic)
plt.show()
```

```python
# hue를 통해 구분 가능
sns.countplot(x='Embarked', hue='Survived', data=titanic)
plt.show()
```

```python
# BarPlot
sns.barplot(x='Embarked', y='Fare', data=titanic)
plt.show()
```

```python
# Heatmap
temp1 = titanic.groupby(['Embarked', 'Pclass'], as_index=False)['PassengerId'].count()
temp2 = temp1.pivot(index='Embarked', columns='Pclass', values='PassengerId')

sns.heatmap(temp2, annot=True, cmap='Blues')
plt.show()
```
